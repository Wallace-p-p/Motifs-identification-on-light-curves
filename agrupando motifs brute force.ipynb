```python
import numpy as np
import pickle
with open('LisMot.txt', 'rb') as fp:
    LM= pickle.load(fp)
```

```python
from biosppy.signals import tools
mot = []
while(len(LM) >= 3 ):
    m=50
    Co= LM[1:]
    C= []
    w=[]
    euclim=[]
    for a in range(len(Co)):
        C= np.concatenate((C, Co[a]), axis=0)
    eucli = tools.distance_profile(query=LM[0], signal=C, metric='euclidean')
    M= []
    M.append(LM[0])
    euc= eucli[0]
    for e in range(len(euc)):
        if((e%m)+len(LM[0])<=m):
            euclim.append(euc[e])
            w.append(e)
    for d in range(len(euclim)):
        if(euclim[d]<(np.min(euclim)+2)):
                u= w[d]
                M.append(C[u:u+m])
                if(len(LM)>int(u/m)+1):
                    del(LM[int(u/m)+1])
    mot.append(M)
    del(LM[0])    
```

```python
import matplotlib.pyplot as plt
r= 9
for j in range(len(mot[r])):
    plt.plot(mot[r][j])
plt.show()
len(mot)
```

```python
l = mot

with open('Mots.txt', 'wb') as fp:
    pickle.dump(l, fp)
with open('Mots.txt', 'rb') as fp:
    U= pickle.load(fp)
    
```
